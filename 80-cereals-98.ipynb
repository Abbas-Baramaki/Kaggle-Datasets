```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import Ridge , LinearRegression
```

```python
data = pd.read_csv('/kaggle/input/80-cereals/cereal.csv')
```

```python
data
```

```python
data.info()
```

```python
data = data.replace(-1,np.nan)
data = data.drop('name',axis=1)
```

```python
missing = ['sugars','carbo','potass']

for mis in missing :
    data[mis] = data[mis].fillna(data[mis].mean())
```

```python
data['type']  = data['type'].apply(lambda x: 1 if x == 'H' else 0)
```

```python
enc_cols = ['mfr']
enc_prefix = ['M_']
```

```python
for col , prefix in zip(enc_cols,enc_prefix):
    dummie = pd.get_dummies(data[col],prefix=prefix).astype(np.int)
    data = pd.concat([data,dummie],axis=1)
    data = data.drop(col,axis=1)
```

```python
y = data['rating']
x = data.drop('rating',axis=1)
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.725,random_state=12)
```

```python
ridge = Ridge(alpha=1.0)
linear = LinearRegression()

ridge.fit(x_train,y_train)
linear.fit(x_train,y_train)

print(ridge.score(x_test,y_test))
print(linear.score(x_test,y_test))
```
