```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
data = pd.read_csv('/kaggle/input/diamonds/diamonds.csv')
```

```python
data.head()
```

```python
Y = data['price']
X = data.drop('price',axis=1)
```

```python
encoder = LabelEncoder()
X['cut'] = encoder.fit_transform(X['cut'])
X['color'] = encoder.fit_transform(X['color'])
X['clarity'] = encoder.fit_transform(X['clarity'])
```

```python
X
```

```python
minmax = MinMaxScaler()
minmax.fit(X)
X = minmax.transform(X)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,train_size=0.33)
```

```python
lin_model = LinearRegression()
lasso_model = Lasso(alpha=0.5,max_iter=1500,tol=100)
ridge_model = Ridge(alpha=0.6,max_iter=1500,tol=100)
```

```python
lin_model.fit(x_train,y_train)
lasso_model.fit(x_train,y_train)
ridge_model.fit(x_train,y_train)
```

```python
print(lin_model.score(x_train,y_train))
print(lin_model.score(x_test,y_test))
```

```python
print(lin_model.score(x_test,y_test))
print(lasso_model.score(x_test,y_test))
print(ridge_model.score(x_test,y_test))
```
