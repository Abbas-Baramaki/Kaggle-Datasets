```python
import numpy as np
import pandas as pd
import tensorflow as tf
import plotly.express as px


from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

dataset url : https://www.kaggle.com/datasets/uciml/indian-liver-patient-records                                 
downlaod : 
https://www.kaggle.com/datasets/uciml/indian-liver-patient-records/download?datasetVersionNumber=1

```python
data = pd.read_csv('/kaggle/input/indian-liver-patient-records/indian_liver_patient.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

## Missing Value's

```python
data[data['Albumin'] == 3.9]
```

```python
data[data['Albumin_and_Globulin_Ratio'].isna()]
```

```python
 data['Albumin_and_Globulin_Ratio'] = data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].mean())
```

## Encoding

```python
 data['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
data['Dataset'] = data['Dataset'].apply(lambda x: 1 if x == 2 else 0)
```

# Splitting And Scaling

```python
y = data['Dataset']
x = data.drop('Dataset',axis=1)
```

```python
y
```

### Scaling

```python
scaler = StandardScaler()

x = pd.DataFrame(scaler.fit_transform(x) , columns = x.columns)
```

#### Spliting

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.75)
```

# Training

```python
y.sum() / len(y)
```

```python
inputs = tf.keras.Input(shape=(10,))
x = tf.keras.layers.Dense(64,activation='relu')(inputs)
x = tf.keras.layers.Dense(64,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs,outputs=outputs)

model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics = [tf.metrics.AUC(name='auc')]
)

epochs = 13
hist = model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=64,
    epochs=epochs,
    callbacks=tf.keras.callbacks.ReduceLROnPlateau(),
    verbose=0

)
```

## Result

```python
fig = px.line(
    hist.history,
    y = ['loss','val_loss'],
    labels={'index' : 'Epochs', 'value' : 'Loss'},
    title= 'Training And Validation Loss'
)
fig.show()
```

```python
np.argmin(hist.history['val_loss']) + 1 
```

```python
s = y.sum() / len(y)
print(model.evaluate(x_test , y_test))
```
