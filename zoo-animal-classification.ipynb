```python
import pandas as pd 
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('/kaggle/input/zoo-animal-classification/zoo.csv')
```

```python
data
```

```python
data['class_type'].unique()
```

```python
len(data['animal_name'].unique())
data.drop('animal_name',axis=1,inplace=True)
```

```python
y = data['class_type']
X = data.drop('class_type',axis=1)
```

```python
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y , train_size=0.85)
```

```python
log_model = LogisticRegression()
svm_model = SVC(C=0.5)
mlp_model = MLPClassifier(hidden_layer_sizes=(18,64,7),validation_fraction=0.2)

log_model.fit(x_train,y_train)
svm_model.fit(x_train,y_train)
mlp_model.fit(x_train,y_train)
```

```python
print('*'*16 + '  Accuracy  ' + '*'*20)
print(f'SVM score : {svm_model.score(x_test,y_test)}\n Logistic score : {mlp_model.score(x_test,y_test)} \n MLP score :  {mlp_model.score(x_test,y_test)}')
```
