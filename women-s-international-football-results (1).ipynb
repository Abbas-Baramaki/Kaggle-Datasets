```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/kaggle/input/womens-international-football-results/results.csv')
```

```python
data
```

```python
sum(data.dtypes == 'object')
```

```python
data['year'] =  data['date'].apply(lambda x: x[0:4]).astype(np.int)
data['month'] = data['date'].apply(lambda x: x[5:7]).astype(np.int)
data = data.drop('date',axis=1)
data['neutral'] = data['neutral'].astype(np.int) 
```

```python
data['away win'] = (data['home_score'] < data['away_score']).astype(np.int)
data = data.drop(['home_score','away_score','home_team','city'],axis=1)
```

```python
nom_cols = ['away_team','tournament','country']
nom_prefix = ['AWAY_','TOURN_','COU_']
```

```python
for col , prefix in zip(nom_cols,nom_prefix):
    dummie = pd.get_dummies(data[col],prefix=prefix).astype(np.int)
    data = pd.concat([data,dummie],axis=1)
    data = data.drop(col,axis=1)
```

```python
y = data['away win']
x = data.drop('away win',axis=1)
```

```python
scaler = StandardScaler()

x = scaler.fit_transform(x)
```

```python
#x.shape
#y.mean()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=7)
```

```python
inputs = tf.keras.Input(shape=(410,))
x = tf.keras.layers.Dense(8,activation='relu')(inputs)
x = tf.keras.layers.Dense(8,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs,outputs=outputs)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics = [tf.keras.metrics.AUC(name='auc')]

)
epochs = 200

hist = model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    validation_data=(x_test,y_test),
    batch_size = 40,
    epochs=epochs,
    callbacks=tf.keras.callbacks.ReduceLROnPlateau()
)
```

```python
fig = px.line(
    hist.history,
    y=['loss','val_loss'],
    labels={'x':'Epochs','y':'Loss'},
    title='training and validation loss'
)
fig.show()
```

```python
np.argmin(hist.history['val_loss']) + 1
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_test = y_test.reshape(-1,1)

y_pred = np.where(y_pred > 0.5 , 1 , 0)

(y_pred == y_test).sum() / len(y_test)
```
