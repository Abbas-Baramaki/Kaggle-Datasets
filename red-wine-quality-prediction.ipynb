```python
import numpy as np
import pandas as pd
import seaborn as sns


from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
data = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

```python
data
```

```python
sns.heatmap(data.corr(),annot=True,vmax=1,vmin=-0.1)
```

```python
fetures_corr = ['alcohol','sulphates','citric acid','fixed acidity']
```

```python
X = data[fetures_corr]
Y = data['quality']
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,train_size=0.85)
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
nn = MLPClassifier(hidden_layer_sizes=(256,256),max_iter = 500 , activation='relu',verbose = 0)
nn.fit(x_train , y_train)
```

```python
forest = RandomForestClassifier(n_estimators=200)
forest.fit(x_train,y_train)
```

```python
print(f'''
SVM SCORE : {svm.score(x_test,y_test)}
MLP SCORE : {nn.score(x_test,y_test)}
FOR SCORE : {forest.score(x_test,y_test)}

''')
```
