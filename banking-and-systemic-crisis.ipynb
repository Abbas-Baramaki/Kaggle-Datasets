```python
import numpy as np
import pandas as pd
import plotly.express as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
data = pd.read_csv('/kaggle/input/africa-economic-banking-and-systemic-crisis-data/african_crises.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop(['case','country'],axis =1,inplace=True)
```

```python
data['banking_crisis'] = data['banking_crisis'].apply(lambda x:1 if x=='crisis' else 0)
```

```python
dummie = pd.get_dummies(data['cc3']).astype(np.float)
data = pd.concat([data,dummie],axis=1)
data.drop('cc3',axis=1,inplace=True)
```

```python
y = data['banking_crisis']
x = data.drop(['banking_crisis'],axis=1)
```

```python
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x))
```

```python
x
```

```python
y.sum() / len(y)#damn
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.7)
```

```python
inputs = tf.keras.Input(23)
x = tf.keras.layers.Dense(128,activation='relu')(inputs)
x = tf.keras.layers.Dense(128,activation='relu')(x)

outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs , outputs=outputs)
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics=[tf.metrics.AUC(name='auc')]

)

epochs = 100

hist = model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=128,
    epochs=epochs,
    callbacks=tf.keras.callbacks.ReduceLROnPlateau(verbose=0),
    verbose=0
)
```

```python
fig = plt.line(
    hist.history,
    y=['loss', 'val_loss'],
    labels={'index': "Epoch", 'value': "Loss"},
    title="Training and Validation Loss"
)

fig.show()
```

```python
model.evaluate(x_test,y_test)
```
