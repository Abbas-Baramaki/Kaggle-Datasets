# Cleaning Data

```python
data['Legendary'] = data['Legendary'].astype(np.int)
data.drop(['Name','Type 2'],axis=1,inplace=True)
data['Type 1'].unique()
```

```python
dummie = pd.get_dummies(data['Type 1']).astype(np.int)
data = pd.concat([data,dummie],axis=1)
data.drop('Type 1',axis=1 , inplace=True)
```

# Selecting Data

```python
y = data['Legendary']
x = data.drop('Legendary',axis=1)
```

```python
y.sum() / len(y)
```

# Scaling Data

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.7,random_state=7)
```

```python
x.shape
```

# Training Data

```python
inputs = tf.keras.Input(shape=(27,))
x = tf.keras.layers.Dense(64,activation='relu')(inputs)
x = tf.keras.layers.Dense(64,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs = inputs , outputs = outputs)

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = [tf.keras.metrics.AUC(name='auc')]

)
epochs = 16

hist = model.fit(
    x_train ,
    y_train , 
    validation_split=0.2,
    batch_size=64,
    epochs=epochs,
    callbacks=tf.keras.callbacks.ReduceLROnPlateau(),
    verbose = 0
         )
```

Result

```python
fig = px.line(
    hist.history,
    y=['loss','val_loss'],
    labels = {'index' : 'Epochs', 'value' : 'Loss'},
    title = 'training and validation loss'
)
fig.show()
```

```python
np.argmin(hist.history['val_loss'])
```

```python
auc_fig = px.line(
    hist.history,
    y=['auc','val_auc'],
    labels = {'index' : 'Epochs' , 'value' : 'AUC Score'},
    title = 'training and validation auc score'
)
auc_fig.show()
```

### Score

```python
model.evaluate(x_test,y_test)
```
