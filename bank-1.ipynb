```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import f1_score
```

```python
data = pd.read_csv('/kaggle/input/bank-direct-marketing/bank.csv',delimiter=';')
```

```python
data
```

```python
dummie_fetures = ['job','education',]
enc_fetures = ['housing','month','loan','default','marital','y','poutcome','contact']
```

```python
data['education'].unique()
```

```python
def prefi(data,column,prefix):
    
    return data[column].apply(lambda x: prefix + x) 
```

```python
data['education'] = prefi(data,'education' , 'e_')
```

```python
data['job'] = prefi(data,'job' , 'j_')
```

```python
#check
data['marital'] = data['marital'].apply(lambda x : 'married' if x == 'married' else 'single')
```

```python
def encod (data,columns):
    encoder = LabelEncoder()
    for col in columns :
        data[col] = encoder.fit_transform(data[col])
    return data
```

```python
def dummie(data,columns):
    for col in columns:
        dummie  = pd.get_dummies(data[col])
        dummie = dummie.astype(np.int)
        data = pd.concat([data,dummie],axis=1)
        data.drop(col,axis=1,inplace=True)
    return data
```

```python
encod(data,enc_fetures)
```

```python
data = dummie(data,dummie_fetures)
```

```python
data.dtypes
```

```python
Y = data['y']
X = data.drop('y',axis=1)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,train_size=0.7)
```

```python
log_model = LogisticRegression()
svm_model = SVC()
nn_model = MLPClassifier(hidden_layer_sizes=(128,128,258,258,2))
```

```python
log_model.fit(x_train,y_train)
svm_model.fit(x_train,y_train)
nn_model.fit(x_train,y_train)
```

```python
print(f'Accuracy : \n MLP : {nn_model.score(x_test,y_test)}\n Logistic : {log_model.score(x_test,y_test)} \n SVM : {svm_model.score(x_test,y_test)}')
```
