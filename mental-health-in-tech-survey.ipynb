```python
import numpy as np
import plotly.express as px
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

####  **https://www.kaggle.com/datasets/osmi/mental-health-in-tech-survey**

```python
data = pd.read_csv('/kaggle/input/mental-health-in-tech-survey/survey.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
cols = {col : data[col].unique() for col in data.columns if data[col].dtypes == 'object'}
```

```python
data['Year'] = data['Timestamp'].apply(lambda x: x[0:4])
data['Month'] = data['Timestamp'].apply(lambda x: x[5:7])
data['Day'] = data['Timestamp'].apply(lambda x: x[8:10])
data['Hour'] = data['Timestamp'].apply(lambda x: x[11:13])
data['Minute'] = data['Timestamp'].apply(lambda x: x[14:16])
data['Second'] = data['Timestamp'].apply(lambda x: x[17:19])
data = data.drop('Timestamp',axis=1)
```

```python
binary_fets = ['Gender','tech_company','treatment','family_history','tech_company']
pos = ['M','Yes','Yes','Yes','Yes']
```

```python
for col , pos in zip(binary_fets,pos):
    data[col] = data[col].apply(lambda x: 1 if x==pos else 0 )
```

```python
def nominal_process(data,cols):
    for col in cols:
        dummie = pd.get_dummies(data[col],prefix=col[0:4]).astype(np.float)
        data = pd.concat([data,dummie],axis=1)
        data = data.drop(col,axis=1)
    return data
```

```python
cols = {col : data[col].unique() for col in data.columns if data[col].dtypes == 'object'}
data = nominal_process(data,cols)
```

```python
data['treatment']
```

```python
y = data['treatment']
x = data.drop('treatment',axis=1)
```

```python
x.shape
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
data['treatment']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.22)
```

```python
inputs = tf.keras.Input(shape=(x_train.shape[1],))
x = tf.keras.layers.Dense(128,activation='relu')(inputs)
x = tf.keras.layers.Dense(128,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs,outputs) 

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics='accuracy'
)

hist = model.fit(
    x_train ,
    y_train ,
    validation_split=0.2,
    batch_size=32,
    epochs=100,
    callbacks = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=7,
        restore_best_weights=True
    )

)
```

```python
pred = np.squeeze(np.array(model.predict(x_test) >= 0.5,dtype=np.int) )
```

```python
model.evaluate(x_test,y_test)
```
