```python
import numpy as np
import pandas as pd
import plotly.express as px
import tensorflow as tf


from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/kaggle/input/covid19-healthy-diet-dataset/Fat_Supply_Quantity_Data.csv')
```

```python
data
```

```python
data.info()
```

```python
data.drop(['Country','Unit (all except Population)'],axis=1,inplace=True)
```

```python
data['Undernourished'].value_counts()
```

```python
data['Undernourished'] = data['Undernourished'].replace('<2.5',0.852)
```

```python
data['Undernourished'] = data['Undernourished'].astype(np.float)
```

# Missing Values 

```python
for col in data.columns:
    if data[col].dtypes != 'object' and data[col].isnull().sum() > 0:
        data[col] = data[col].fillna(data[col].mean())
```

# Selecting Target

```python
data['Deaths'] = pd.qcut(data['Deaths'] , q=2 , labels=[0,1])
target ='Deaths'
data[target].unique()
```

```python
y = data[target]
x = data.drop(target,axis=1)
```

```python
data.describe()
```

# Scaling

```python
scaler = MinMaxScaler()

x = pd.DataFrame(scaler.fit_transform(x) , columns = x.columns)
x
```

# Training and spliting

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.7,random_state=7)
```

```python
(y == 1).sum() / len(y)
```

```python
inputs = tf.keras.Input(shape=(29,))
x = tf.keras.layers.Dense(64,activation='relu')(inputs)
x = tf.keras.layers.Dense(64,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)
model = tf.keras.Model(inputs=inputs,outputs=outputs)

model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics=['accuracy',tf.keras.metrics.AUC(name='auc')]
)
epochs=180
hist = model.fit(
    x_train,
    y_train,
    validation_split=0.3,
    batch_size=64,
    epochs=epochs,
    #callbacks=tf.keras.callbacks.ReduceLROnPlateau(),
    verbose=0
)
```

# Result

```python
fig = px.line(
    hist.history,
    y=['loss','val_loss'],
    labels={'index' : 'Epochs' , 'value' : 'Loss'}
   ,title = 'Validation And Training Loss'
)
fig.show()
```

```python
np.argmin(hist.history['val_loss']) +1 
```

```python
model.evaluate(x_test,y_test)
```
