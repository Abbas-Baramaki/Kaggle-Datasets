```python
import pandas as pd
import numpy as np
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
data = pd.read_csv('/kaggle/input/top-personality-dataset/2018-personality-data.csv')
```

```python
data
```

```python
data.info()
```

```python
data.drop('userid',axis=1,inplace=True)
```

```python
data[' assigned metric'].unique()
```

```python
data[' assigned condition'].unique()
assigned_condition_ordinal = [ ' low',' medium',' default',' high']
```

```python
dummie = pd.get_dummies(data[' assigned metric']).astype(np.float)
data = pd.concat([data,dummie],axis=1)
data.drop(' assigned metric',axis=1,inplace=True)
```

```python
data[' assigned condition'] = data[' assigned condition'].apply(lambda x : assigned_condition_ordinal.index(x)) 
```

```python
data.describe()
```

```python
y = data[' enjoy_watching ']
x = data.drop(' enjoy_watching ',axis=1)
```

```python
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
x_train , x_test , y_train  ,y_test = train_test_split(x,y,train_size=0.7,random_state=7)
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (64,64))
svm_model  = SVC()
log_model = LogisticRegression()

mlp_model.fit(x_train,y_train)
svm_model.fit(x_train,y_train)
log_model.fit(x_train,y_train)
```

```python
print(f'nn score : {mlp_model.score(x_test , y_test)}\n svm score : {svm_model.score(x_test , y_test)}\n log score : {log_model.score(x_test , y_test)}')
```

```python
fig = px.bar(
 x= ['Log Accuracy' , 'Support Vector Accuracy' , 'MLP Accuracy'],
 y = [log_model.score(x_test ,y_test),svm_model.score(x_test ,y_test),mlp_model.score(x_test ,y_test)],
title = 'Three models accuracy compare'
)
fig.show()
```
