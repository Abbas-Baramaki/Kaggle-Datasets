```python
import numpy as np
import pandas as pd

import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data  = pd.read_csv('/kaggle/input/white-wine-quality/winequality-white.csv',delimiter=';')
```

```python
data
```

```python
data.info()
```

# Encoding

```python
target = 'quality'
ordined = sorted((list(data[target].unique())))
data[target] = data[target].apply(lambda x: ordined.index(x))
```

```python
data[target].unique()
```

# Scaling , Selecting And Splitting

```python
y = data[target]
x = data.drop(target,axis=1)
```

```python
y = pd.qcut(y,q=2,labels=[0,1])
```

```python
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x),columns = x.columns)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.75,random_state=7)
```

```python
y.astype(np.float).sum() / len(y)
```

# Training Model

```python
inputs = tf.keras.Input(shape=(11,))
x = tf.keras.layers.Dense(64,activation='relu')(inputs)
x = tf.keras.layers.Dense(64,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs = inputs , outputs = outputs)

model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics = [tf.keras.metrics.AUC(name='auc')]
)
epochs=29

hist = model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=64,
    epochs=epochs,
    callbacks=tf.keras.callbacks.ReduceLROnPlateau(),
    verbose=0
)
```

# Result

```python
fig = px.line(
    hist.history,
    y = ['loss','val_loss'],
    labels = {'x' : 'Epochs' , 'y' : 'Loss'},
    title = 'training and validation loss'

)
fig.show()
```

```python
np.argmin(hist.history['val_loss']) + 1
```

```python
model.evaluate(x_test,y_test)
```
