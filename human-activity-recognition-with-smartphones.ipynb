```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
train_data = pd.read_csv('/kaggle/input/human-activity-recognition-with-smartphones/train.csv')
test_data = pd.read_csv('/kaggle/input/human-activity-recognition-with-smartphones/test.csv')
```

```python
encoder = LabelEncoder()
scaler = StandardScaler()
y_train = encoder.fit_transform(train_data['Activity'])
y_train_classes = dict(enumerate(encoder.classes_))
x_train = train_data.drop('Activity',axis=1)
x_train = scaler.fit_transform(x_train)

scaler_2 = StandardScaler()
encoder_2 = LabelEncoder()
y_test = encoder_2.fit_transform(test_data['Activity'])
y_test_classes = dict(enumerate(encoder_2.classes_))
x_test = test_data.drop('Activity',axis=1)
x_test = scaler_2.fit_transform(x_test)
```

```python
train_data['Activity'].unique()
```

```python
inputs = tf.keras.Input(shape=(x_train.shape[1]))
x = tf.keras.layers.Dense(128,activation='relu')(inputs)
x = tf.keras.layers.Dense(128,activation='relu')(x)
outputs = tf.keras.layers.Dense(6,activation='softmax')(x)
                          
model = tf.keras.Model(inputs,outputs)
                                
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics='accuracy'
)  
hist = model.fit(
    x_train,
    y_train,
    validation_split =0.12,
    batch_size=40,
    epochs=50,
    callbacks=[
        tf.keras.callbacks.ModelCheckpoint(r'/kaggle/working/Human-Activity-Recognition-with-Smartphones.h5'
                                           ,save_best_only=True),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_accuracy',
            patience=3
        )
        
    ]

)
```

```python
model.load_weights('/kaggle/working/Human-Activity-Recognition-with-Smartphones.h5')
```

```python
model.evaluate(x_test,y_test)
```
