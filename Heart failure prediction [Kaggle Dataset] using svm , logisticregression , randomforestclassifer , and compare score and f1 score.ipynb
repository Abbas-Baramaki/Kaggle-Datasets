```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import f1_score
from pandas_profiling import ProfileReport
from sklearn.model_selection import train_test_split 

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
data = pd.read_csv('/kaggle/input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
data
```

```python
profile = ProfileReport(data)
```

```python
profile
```

```python
Y = data['DEATH_EVENT']
X = data.drop('DEATH_EVENT',axis=1)
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,train_size=0.65)
```

```python
log_model = LogisticRegression(verbose=True,penalty='l1')
svm_model = SVC()
rand_model = RandomForestClassifier()
```

```python
log_model.fit(x_train,y_train)
y_pred = log_model.predict(x_test)
score_log = log_model.score(x_test,y_test)
scoref_log = f1_score(y_pred,y_test)
```

```python
svm_model.fit(x_train,y_train)
y_pred = svm_model.predict(x_test)
score_svm = svm_model.score(x_test,y_test)
scoref_svm = f1_score(y_pred,y_test)
```

```python
rand_model.fit(x_train,y_train)
y_pred = rand_model.predict(x_test)
score_rand = rand_model.score(x_test,y_test)
scoref_rand = f1_score(y_pred,y_test)
```

```python
print(f"Score Model Log: {score_log} ,\t F1Score {scoref_log}\n")
print(f"Score Model Log: {score_svm} ,\t F1Score {scoref_svm}\n")
print(f"Score Model Log: {score_rand} ,\t F1Score {scoref_rand}\n")
```
