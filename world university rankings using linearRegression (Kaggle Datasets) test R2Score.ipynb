```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/kaggle/input/world-university-rankings/cwurData.csv')
```

```python
data1 = pd.read_csv('/kaggle/input/world-university-rankings/cwurData.csv')
```

```python
data.drop('institution',axis=1,inplace=True)
```

```python
data.drop('broad_impact',axis=1,inplace=True)
```

```python

encoder = LabelEncoder()
data['country'] = encoder.fit_transform(data['country'])
```

```python
X = data.drop('world_rank',axis=1)
Y = data['world_rank']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
x_train , x_test , y_train , y_test  = train_test_split(X,Y,train_size=0.8)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)

print(model.score(x_test,y_test))
```
