```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from pandas_profiling import ProfileReport
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
data
```

```python
encoder = LabelEncoder()
data['gender'] = encoder.fit_transform(data['gender'])
gender = {index:label  for index , label in enumerate(encoder.classes_)}

data['race/ethnicity'] = encoder.fit_transform(data['race/ethnicity'])
race = {index:label for index,label in enumerate(encoder.classes_)}

data['lunch'] = encoder.fit_transform(data['lunch'])
lunch = {index:label for index,label in enumerate(encoder.classes_)}

data['test preparation course'] = encoder.fit_transform(data['test preparation course'])
prep = {index:label for index,label in enumerate(encoder.classes_)}

data['parental level of education'] = encoder.fit_transform(data['parental level of education'])
parental = {index:label for index,label in enumerate(encoder.classes_)}
```

```python
print(f'1\t{gender} \n 2\t{race} \n 3\t{lunch} \n 4\t{prep} \n 5\t{parental}')
```

```python
report = ProfileReport(data)
```

```python
scaler = StandardScaler()
```

```python
x_math = data[['test preparation course']]
x_math = scaler.fit_transform(x_math)
y_math = data[['math score']]
y_math = scaler.fit_transform(y_math)

x_reading = data[['writing score']]
x_reading = scaler.fit_transform(x_reading)
y_reading = data[['reading score']]
y_reading = scaler.fit_transform(y_reading)
```

```python
x_math_train , x_math_test , y_math_train , y_math_test = train_test_split(x_math , y_math,train_size=0.70)
x_reading_train , x_reading_test , y_reading_train , y_reading_test = train_test_split(x_reading , y_reading , train_size = 0.70)
```

```python
math_model = LinearRegression()
math_model.fit(x_math_train,y_math_train)

reading_model = LinearRegression()
reading_model.fit(x_reading_train,y_reading_train)
```

```python
R2 = math_model.score(x_math_test,y_math_test)
R2_reading = reading_model.score(x_reading_test,y_reading_test)
```

```python
print(f'R2 Score Math is  {R2}\n R2 Score Reading is {R2_reading}')
```
