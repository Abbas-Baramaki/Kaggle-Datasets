```python
import numpy as np
import tensorflow as tf
```

```python
training_options = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    rotation_range=50,
    width_shift_range=0.5,
    height_shift_range=0.5,
    validation_split=0.2
)
train = training_options.flow_from_directory(
    '/kaggle/input/gender-detection-and-classification-image-dataset/train'
    ,target_size=(224,224)
    ,class_mode='binary'
    ,batch_size=10
)


valid_options = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    rotation_range=50,
    width_shift_range=0.5,
    height_shift_range=0.5,
    validation_split=0.2
)
valid = training_options.flow_from_directory(
    '/kaggle/input/gender-detection-and-classification-image-dataset/train'
    ,target_size=(224,224)
    ,class_mode='binary'
    ,batch_size=10
)


test_options = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale = 1./255
)
test = test_options.flow_from_directory(
    '/kaggle/input/gender-detection-and-classification-image-dataset/test'
    ,target_size=(224,224)
    ,class_mode='binary'
    ,batch_size=10
)
```

```python
inputs = tf.keras.Input(shape = (224,224,3))
#layer 1
x = tf.keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu')(inputs)
x = tf.keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(x)
#layer 2
x = tf.keras.layers.Conv2D(128,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(128,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(x)
#layer 3
x = tf.keras.layers.Conv2D(256,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(256,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(256,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(x)
#layer 4
x = tf.keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(x)
#layer 5
x = tf.keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.Conv2D(512,kernel_size=(3,3),activation='relu')(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(x)

x = tf.keras.layers.Flatten()(x)

x = tf.keras.layers.Dense(4096,activation='relu')(x)
x = tf.keras.layers.Dense(1000,activation='relu')(x)
outputs = tf.keras.layers.Dense(1,activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs,outputs=outputs)

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy']

)

epochs = 4

with tf.device('GPU:0'):
    hist = model.fit(
        train,
        validation_data=valid,
        batch_size=20,
        epochs=epochs,
        callbacks = tf.keras.callbacks.EarlyStopping(
            monitor='val_loss'
            ,patience=3
            ,restore_best_weights=True
            ,verbose = 1
        )

    )
```

```python
model.evaluate(test)
```
